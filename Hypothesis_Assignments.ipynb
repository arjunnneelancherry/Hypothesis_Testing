NULL HYPOTHESIS : The diameter of the cutlet between two units are significant

ALTERNATE HYPOTHESIS :There is no significant diameter of the cutlet between two units

```python
stats.f_oneway(data['Unit A'],data['Unit B'])
```

if the p-value is less than 0.05 Alternate Hypothesis is accepted if it is greater than 0.05 so Null Hypothesis will be accepted.

In here pvalue is lesser than 0.05 that means Alternate Hypothesis will be accepted

```python
import pandas as pd
from scipy import stats
```

```python
from google.colab import files
uploaded= files.upload()
```

```python
data=pd.read_csv('Costomer+OrderForm.csv')
data
```

Null Hypothesis : The defective % will not varies by centre

Alternate Hypothesis : The defective % will varies by centre

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['Phillippines']=le.fit_transform(data['Phillippines'])
data['Indonesia']=le.fit_transform(data['Indonesia'])
data['Malta']=le.fit_transform(data['Malta'])
data['India']=le.fit_transform(data['India'])
```

```python
data.head()
```

```python
stats.f_oneway(data['Phillippines'],data['Indonesia'],data['Malta'],data['India'])
```

Since pvalue is lesser than 0.05 then Alternate Hypothesis
